```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
countries = pd.read_csv("countries.csv")
countries.head()
```

```python
china = countries[countries.country == "China"]
usa = countries[countries.country == "United States"]
```

```python
plt.title("Population Growth Comparision (Absolute)")
plt.plot(china.year, china.population / 1e6)
plt.plot(usa.year, usa.population / 1e6)
plt.ylabel("Population (Millions)")
plt.xlabel("Year")
plt.legend(["China", "USA"])
plt.show()
```

```python
china_popgrowth = china.population / china.population.iloc[0] * 100
usa_popgrowth = usa.population / usa.population.iloc[0] * 100
```

```python
plt.title("Population Growth Comparision (Relative)")
plt.plot(china.year, china_popgrowth)
plt.plot(usa.year, usa_popgrowth)
plt.ylabel("Population growth (Base = 100)")
plt.xlabel("Year")
plt.legend(["China", "USA"])
plt.show()
# More meaningful representation.
```
