```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("countries.csv")
data.head()
```

```python
mean_gdpPerCapita = data.groupby(["country"]).mean().gdpPerCapita
top10gdpPerCapita = mean_gdpPerCapita.sort_values(ascending = False).head(10)
top10gdpPerCapita.head(5)
```

## Is Kuwait really the richest country in the world (in GDP per capita)?

```python
kuwait = data[data.country == "Kuwait"]
kuwait.head()
```

```python
plt.title("Change in GDP per capita in Kuwait over time")
plt.plot(kuwait.year, kuwait.gdpPerCapita)
plt.xlabel("Year")
plt.ylabel("GDP per capita")
plt.show()
```

```python
plt.subplot(311)
plt.title("GDP per capita in millions")
plt.plot(kuwait.year, kuwait.gdpPerCapita / 1e6)

plt.subplot(312)
plt.title("GDP in billions")
plt.plot(kuwait.year, kuwait.gdpPerCapita * kuwait.population / 1e9)

plt.subplot(313)
plt.title("Population in billions")
plt.plot(kuwait.year, kuwait.population / 1e6)

plt.tight_layout()
plt.show()
```

```python
plt.title("Relative comparision of GDP and GDP per capita with population")
plt.plot(kuwait.year, kuwait.population / kuwait.population.iloc[0] * 100)

kuwait_gdp = kuwait.gdpPerCapita * kuwait.population
plt.plot(kuwait.year, kuwait_gdp / kuwait_gdp.iloc[0] * 100)
plt.plot(kuwait.year, kuwait.gdpPerCapita / kuwait.gdpPerCapita.iloc[0] * 100)
plt.legend(["Population", "GDP", "GDP per capita"])
plt.xlabel("Year")
plt.show()
```

## Thus Kuwait cannot be said to be the richest country with absolute certainity, as their GDP per capita has decreased over time.

```python
plt.title("Comparision between Norway and Kuwait")
plt.plot(norway.year, norway.gdpPerCapita)
plt.plot(kuwait.year, kuwait.gdpPerCapita)
plt.xlabel("Year")
plt.ylabel("GDP per capita")
plt.show()
```

### This proves the above claim by showing that currently Norway has a higher GDP per capita than Kuwait. Thus the richest country in the world cannot be clearly determined by this data.
