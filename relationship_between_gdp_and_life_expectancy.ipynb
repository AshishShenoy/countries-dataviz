```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("countries.csv")
data.head()
```

```python
data_2007 = data[data.year == 2007]
data_2007.head()
```

```python
# Using absolute GDP per capita.
plt.title("Relation between GDP per capita and life expectancy in 2007")
plt.scatter(data_2007.gdpPerCapita, data_2007.lifeExpectancy, 8)
plt.xlabel("GDP per capita")
plt.ylabel("Life expectancy")
plt.show()
```

```python
data_2007.lifeExpectancy.corr(data_2007.gdpPerCapita)
```

```python
# Using the logarithm of the GDP per capita.
plt.title("Relation between GDP per capita and life expectancy in 2007")
plt.scatter(np.log10(data_2007.gdpPerCapita), data_2007.lifeExpectancy, 8)
plt.xlabel("GDP per capita")
plt.ylabel("Life expectancy")
plt.show()
# MOre linear relation.
```

```python
data_2007.lifeExpectancy.corr(np.log10(data_2007.gdpPerCapita))
# Higher correlation.
```

```python
# Linear Scale
years = sorted(set(data.year))
for given_year in years:
    data_year = data[data.year == given_year]
    plt.title(given_year)
    plt.scatter(data_year.gdpPerCapita, data_year.lifeExpectancy, 5)
    plt.xlim(0, 60000)
    plt.ylim(30, 80)
    plt.xlabel("GDP per capita")
    plt.ylabel("Life expectancy")
    plt.show()
```

```python
# Log Scale
years = sorted(set(data.year))
for given_year in years:
    data_year = data[data.year == given_year]
    plt.title(given_year)
    plt.scatter(np.log10(data_year.gdpPerCapita), data_year.lifeExpectancy, 5)
    plt.xlim(2, 5)
    plt.ylim(30, 80)
    plt.xlabel("GDP per capita")
    plt.ylabel("Life expectancy")
    plt.show()
```
